```python
# checking if "pdf" is contained in "href" so we can drop href

counter = []
for i in range(len(df)):
    if any(df['href'].str.contains(df['pdf'][i])):
        counter.append(df['pdf'][i])

len(counter) == len(df)

# all pdf names are contained in the corresponding href column so we can drop the href column safely
```

```python
# check if the Case Number is contained in the pdf file name
cn_counter = []
for i in range(len(df)):
    if any(df['pdf'].str.contains(df['Case Number'][i])):
        cn_counter.append(df['Case Number'][i])

len(cn_counter) / len(df)

# ~ 98% of the case numbers are contained in the pdf name so we can safely drop the case number column
```

```python
# check if Case Number.1 and Case Number.2 are identical before dropping them
identical_counter = []
for i in range(len(df)):
    if any(df['Case Number.2'].str.contains(df['Case Number.2'][i])):
        identical_counter.append(df['Case Number.2'][i])

len(identical_counter) / len(df)

# Case Number.1 is identical to Case Number.2 so we can drop both columns since we already know that
# the case number is contained in the pdf file name which we are keeping in the data set
```

```python
# checking if "original order" column contains related numbers starting from 0 so we can use the column as
# an index column
row_count = len(df)
original_order_values = df["original order"].nunique()
print(row_count == original_order_values)

# the value count for unique values in "original order" column does not match the length of 
# the data set. this means that there are either duplicate or missing values in the "original order" column 
# so we won't use is as an index column
```

```python
# What I am going to do to clean this data set:

# removing duplicate and unrelated information and renaming columns
# 1. removing spaces after the labels "Sex" and "Species" on "Sex " and "Species " columns
# 2. remove the "Case Number" column because the information is already stored in the "pdf" column inside 
# the file name
# 3. removing columns "Case Number.1" and "Case Number.2" because they are duplicates from "Case Number"
# 4. removing columns "Unnamed: 22" and "Unnamed: 23" because they don't contain meaningful information
# 5. removing columns "href formula" and "href" because of duplicate information to "pdf" - the pdf files
# can still be found via Google by file name and most of the links in href are dead anyways
# 6. removing "year" column because the information is already present in the date column
# 7. removing column "original order" because it does not contain meaningful information and can't be 
# used as an index column

# cleaning values in columns
# 1. remove non numeric values from "Age" column
# 2. removing unrelated/non activity values from "Activity" column
# 3. removing quoting characters from values in "Activity" column
# 4. removing non name entries from the "Name" column: "2 males", "male", non alphabetical values, etc.
# 5. capitalize values in the columns "Name", "Location", "Area", "Activity", "Injury", "Species"

# reshaping and combining columns
# 1. sum up "M", "F" and "O(ther)" values in "Sex" table
# 2. combining "Date" and "Time" column into a "Datetime" column, replacing/processing missing values
# 3. group the values "Boat" and "Boating" in the "Type" column into "Boating" because the meaning should 
# be the same
# 4. change "Invalid" values into np.nan values inside of "Type" column

# changing dtypes of columns
# 1. change value types in "Age" column to float
# 2. replacing "Fatal (Y/N)" with boolean values
# 3. converting new "Datetime" column into datetime object

# processing entire data set
# 1. check for duplicate rows
# 2. check for rows with less than 4 valid values inside
# 3. writing cleaned data to a csv file

# what the data set's data types should look like after the clean up:
# Datetime                  datetime
# Type                      object
# Country                   object
# Area                      object
# Location                  object
# Activity                  object
# Name                      object
# Sex                       object
# Age                       float
# Injury                    object
# Fatal (Y/N)               bool
# Species                   object
# Investigator or Source    object
# pdf                       object

```

```python
# checking for duplicate rows 
duplicate_rows = df[df.duplicated(keep='first')]
duplicate_rows

# there are no duplicate rows we have to process
```

```python
# removing space after the labels "Sex" and "Species" on "Sex " and "Species " columns for improved
# processability (and less key errors)

df = df.rename(columns={"Sex ": "Sex", "Species ": "Species"})
df.head()
```

```python
# removing columns "Case Number.1" and "Case Number.2" as well as "Case Number"
# removing columns "Unnamed: 22" and "Unnamed: 23"
# removing columns "href formula" and "href" 
# removing column "year"
# removing column "original order" 

df = df.drop(["Case Number.1", "Case Number.2", "Unnamed: 22", "Unnamed: 23", "href formula", "href", "Case Number", "Year", "original order"], axis=1)
df
```

```python
# remove non numeric values from "Age" column
df["Age"] = pd.to_numeric(df["Age"], errors='coerce')
df.groupby("Age").nunique().head()
```

```python
# removing unrelated values from "Activity" column
invalid_entries = [".", ",", " ", "  ", "   ", "    ", "male"]
for entry in invalid_entries:
    df = df[df["Activity"] != entry]

df.groupby("Activity").nunique()
df["Activity"].value_counts().head()
```

```python
# removing quotes from values in "Activity" column to improve 'processability'
df["Activity"] = df["Activity"].str.replace("\"", "").str.replace("\'", "")
df.groupby("Activity").nunique()
```

```python
# removing non name entries from the "Name" column: numeric values
df["Name"] = np.where(df["Name"].str.contains("[0-9]+", regex=True), np.nan, df["Name"])

# removing quotes from values in "Name" column to improve 'processability'
df["Name"] = df["Name"].str.replace("\"", "").str.replace("\'", "")

# capitalizing name column now in order to simplify the replacing process
df["Name"] = df["Name"].str.title()

# replacing the most common non name values with nan: "Male", "Female", "Boy", "Girl", etc. is present in the
# "Sex" column and can be replaced with nan, "Unknown" and "Anonymous" should mean the same and can be
# replaced with nan
names_to_replace = ["Male", " Male", "Female", "Males", "Girl", "Boat", "Anonymous", "Unknown", "Woman", "Man", "Boy"]
for name in names_to_replace:
    df["Name"] = np.where(df["Name"] == name, np.nan, df["Name"])

df["Name"].value_counts().head()
```

```python
# sum up "M", "F" and "O(ther)" values in "Sex" table
df["Sex"] = np.where(df["Sex"] == "M ", "M", df["Sex"])
df["Sex"] = np.where((df["Sex"] != "M") & (df["Sex"] != "F") & (df["Sex"]), "O", df["Sex"])
df.groupby("Sex").nunique()
```

```python
# group the values "Boat" and "Boating" in the "Type" column into "Boating"
df["Type"] = np.where((df["Type"] == "Boat") | (df["Type"] == "Boating"), "Boating", df["Type"])
# changing "Invalid" values to nan
df["Type"] = np.where(df["Type"] == "Invalid", np.nan, df["Type"])
df.groupby("Type").nunique()
```

```python
# change date column to datetime object
# removing string prefixes
df["Date"] = df["Date"].str.replace("Reported ", "")

# replace common non convertable date formats
dates_to_replace = {"16-Aug--2011": "16-Aug-2011",
                   "11-Aug--2011": "11-Aug-2011",
                   "190Feb-2010": "19-Feb-2010",
                   "Sep-08": "01-Sep-08"}

for date, replacement in dates_to_replace.items():
    df["Date"] = np.where(df["Date"] == date, replacement, df["Date"])

# convert date string into datetime object to get rid of values which can not be converted to datetime e.g. "World War II"
df["Date"] = pd.to_datetime(df["Date"], errors = "coerce")
# convert datetime object back to string to be able to concat it with time string in the next step
df["Date"] = np.where(df["Date"] != pd.NaT, df["Date"].dt.strftime('%d-%m-%Y'), df["Date"])

df["Date"].value_counts().head()
```

```python
# process "Time" column: convert most common general time spans e.g. "Afternoon" into hh:mm format by 
# assuming mean time point of each period, convert these values to dt.time object and replacing 
# the rest with pd.NaT

times_to_replace = {"Afternoon": "15h30", 
                    "After noon": "15h30", 
                    "Midnight": "00h00", 
                    "Morning": "09h00",
                    "Morning ": "09h00", 
                    "Night": "02h00", 
                    "Evening": "19h00", 
                    "Late afternoon": "17h00", 
                    "Sometime between 06:00 & 08:oo": "07h00", 
                    "Early afternoon": "14h00", 
                    "07:00 - 08:00": "07h30", 
                    "09:00 - 09:30": "09h15", 
                    "830": "08h30", 
                    "Just before noon": "11h30", 
                    "Early morning": "06h00", 
                    "Dawn": "19h30",
                    "Midday": "12h00",
                    "AM": "10h00",
                    "A.M.": "10h00",
                    "Dusk": "06h00",
                    "Lunchtime": "12h00",
                    "15j45": "15h45",
                    "500": "05h00",
                    "Before 07:00": "06h30",
                    "10:00 -- 11:00": "10h30",
                    '"Just before 11:00"': "10h45"}

for span, replacement in times_to_replace.items():
    df["Time"] = np.where(df["Time"] == span, replacement, df["Time"])

# replacing time seperator "h" with ":"
df["Time"] = df["Time"].str.replace("h", ":")

# combining columns "Date" and "Time" into a "Datetime" column
df["Datetime"] = np.where(df["Date"] != "NaT", (df["Date"] + "-" + df["Time"]), df["Date"])

#df["Datetime"] = df["Date"] + "-" + df["Time"]
df["Datetime"] = pd.to_datetime(df["Datetime"], errors = "coerce")
df["Datetime"].value_counts().head()

# dropping the "Date" and "Time" columns because the information is now combined into a datetime object in
# the "Datetime" column
df = df.drop(["Date", "Time"], axis=1)
df.head()
```

```python
# replacing "Fatal (Y/N)" with boolean values
# first clean up the options
df["Fatal (Y/N)"] = np.where((df["Fatal (Y/N)"] == "N ") | (df["Fatal (Y/N)"] == "n") | (df["Fatal (Y/N)"] == " N"), "N", df["Fatal (Y/N)"])
df["Fatal (Y/N)"] = np.where((df["Fatal (Y/N)"] == "#VALUE!") | (df["Fatal (Y/N)"] == "UNKNOWN") | (df["Fatal (Y/N)"] == "F"), np.nan, df["Fatal (Y/N)"])

# then change column type and assign new values according to previous value
df["Fatal (Y/N)"] = np.where(df["Fatal (Y/N)"] == "Y", True, df["Fatal (Y/N)"])
df["Fatal (Y/N)"] = np.where(df["Fatal (Y/N)"] == "N", False, df["Fatal (Y/N)"])
df["Fatal (Y/N)"] = df["Fatal (Y/N)"].astype(bool)
df.dtypes
```

```python
# capitalize values in the columns "Location", "Area", "Activity", "Injury", "Species"
values_to_capitalize = ["Injury"]
values_to_title = ["Location", "Area", "Species", "Activity"]

# capitalizing eyery word in "Location", "Area", "Species", "Activity"
for col in values_to_title:
    df[col] = df[col].str.title()

# capitalizing first word in "Injury"
for col in values_to_capitalize:
    df[col] = df[col].str.capitalize()

df.head(10)
```

```python
# dropping rows with less than 4 meaningful values out of 14 (< 30% of meaningful values)

# first replacing "NaN" and "NaT" values of string type with np.nan in case there are any string values present
df.replace(["NaN", 'NaT'], np.nan, inplace = True)

# then dropping all rows containing less than 5 non NaN values because these rows 
df = df.dropna(thresh=4)
```

```python
# writing cleaned data into new csv file
df.to_csv("GSAF5_clean.csv")
```
